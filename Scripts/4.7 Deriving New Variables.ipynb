## Assign flags for price ranges

### if-statement with user-defined function

```python
#define new function called price_label

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
# apply new function/column

df['price_range'] = df.apply(price_label, axis=1)
```

axis=1 tells Python to apply to all rows vs columns which would be axis = 0

note warning - explore loc() in next section

```python
df['price_range'].value_counts(dropna=False)
```

There are no products greater that $15 in this subset
curious? check most expensive product

```python
df['prices'].max()
```

### Explore if-statements with loc() function

can apply conditional logic of an if-statement without explicitly doing so 

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna=False)
```

Even though result is same, loc() runs faster because loc() function applies the conditional filters before searching through the data frame and Python thinks its better with no warning

```python
ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merged.loc[(ords_prods_merged['prices'] <= 15) & (ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merged['price_range_loc'].value_counts(dropna=False)
```

No memory issues with loc() vs user defined function

### If-statements with For-Loops

We will create a new column that summarizes how busy each day of the week is

```python
## frequency of oders each day of the week
ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

value 0 = Saturday. Saturday has the highest orders, Sunday(1) is the second highest, Wednesday(4) is the slowest

```python
## for loop code to creat new column
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

result = [] creates an empty list which the for-loop then fills in

why this vs. user defined function

you're only looping through one column of df ---> faster performance

```python
result
```

Not yet combined with ords_prods_merged df yet. must create a column within df and set it equal to "results"

```python
## creating column within df
ords_prods_merged['busiest_day']=result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna=False)
```

## Create a Busiest Days Column

```python
## for loop code
new_result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value in [0, 1]:
    new_result.append("Busiest day")
  elif value in [3, 4] :
    new_result.append("Least busy")
  else:
    new_result.append("Regularly busy")
```

```python
new_result
```

```python
## creating new column within df
ords_prods_merged['busiest_days']=new_result
```

```python
ords_prods_merged['busiest_days'].value_counts(dropna=False)
```

```python
# delete busiest day
ords_prods_merged=ords_prods_merged.drop(columns=['busiest_day'])
```

## Find Busiest hours 

```python
## frequency of oders each hour
ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
## for loop code
hours = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [0, 1]:
    hours.append("Most orders")
  elif value in [4, 3]:
    hours.append("Fewest orders")
  else:
    hours.append("Average orders")
```

```python
hours
```

```python
## creating new column within df
ords_prods_merged['busiest_period_of_day']=hours
```

```python
#Frequency
ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```
