## 0.3 Aggregate data with transform()

### step 3. adds a flag

Create a loyalty flag column. First we use transform() for order number frequency. then use loc() to create a loyal or not flag.
If customer maximum orders is > 40 label as "loyal customer', >10 <=40 'regular customer', <= 10 'new customer"

```python
# create max order column using the 3 steps above (3rd step is first)
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check accuracy
ords_prods_merge.head(100)
```

# 03. Deriving Columns with loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

### Step 5: check spending habits of 3 types of customers

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','median','min','max']})
```

### Step 6: create a spending flag column

```python
# create average_price column
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check accuracy
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'average_price']].head(30)
```

### Step 7: create a frequency column

```python
# create a frequency column
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check accuracy
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] <= 20) & (ords_prods_merge['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id', 'frequency_flag', 'days_since_prior_order]].head(60)
```
